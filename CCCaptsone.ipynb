```python
# Merge the dataframes on the 'ID' column
df = pd.merge(customer_demographics, customer_behavior, on='ID')
df = pd.merge(df, credit_consumption, on='ID')

print(df.info(),'\n')
# Display the first few rows of the merged dataframe
print("Merged Dataframe:")
df
```

```python
# Step 3: Data Preparation for Modeling

# Check for missing values in the dataframe
missing_values = df_not_missing.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Fill missing values with the median for numeric columns
for column in df_not_missing.select_dtypes(include=['float64', 'int64']).columns:
    df_not_missing[column].fillna(df_not_missing[column].median(), inplace=True)

# Fill missing values with the mode for categorical columns
for column in df_not_missing.select_dtypes(include=['object']).columns:
    df_not_missing[column].fillna(df_not_missing[column].mode()[0], inplace=True)
```

```python
# Verify that there are no missing values left
print("Missing values after filling:\n", df_not_missing.isnull().sum())
```

```python
# One-hot encode categorical variables
df_encoded = pd.get_dummies(df_not_missing)

# Display the first few rows of the encoded dataframe
print("Encoded dataframe:\n") 
print(df_encoded.shape)
df_encoded.head()
```

```python
# Remove the 'ID' column from features
df_encoded = df_encoded.drop(columns=['ID'])

# Display the first few rows of the dataframe after removing 'ID'
print("Dataframe after removing 'ID':\n") 
print(df_encoded.shape)
df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df_encoded.drop(columns=['cc_cons'])
y = df_encoded['cc_cons']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Display the shapes of the training and test sets
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform both training and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Display the shape of the scaled training and test sets
print("Scaled training set shape (MinMaxScaler):", X_train_scaled.shape)
print("Scaled test set shape (MinMaxScaler):", X_test_scaled.shape)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Define a custom RMSPE metric
def rmspe(y_true, y_pred):
    return np.sqrt(np.mean(np.square((y_true - y_pred) / y_true)))

# Create a Linear Regression model
lr = LinearRegression()

# Train the model
lr.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_lr = lr.predict(X_test_scaled)

# Calculate RMSPE for Linear Regression
rmspe_lr = rmspe(y_test, y_pred_lr)
print(f'RMSPE for Linear Regression: {rmspe_lr:.4f}')
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regressor model
rf = RandomForestRegressor(random_state=42)

# Train the model
rf.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_rf = rf.predict(X_test_scaled)

# Calculate RMSPE for Random Forest Regressor
rmspe_rf = rmspe(y_test, y_pred_rf)
print(f'RMSPE for Random Forest Regressor: {rmspe_rf:.4f}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Create a Gradient Boosting Regressor model
gbr = GradientBoostingRegressor(random_state=42)

# Train the model
gbr.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_gbr = gbr.predict(X_test_scaled)

# Calculate RMSPE for Gradient Boosting Regressor
rmspe_gbr = rmspe(y_test, y_pred_gbr)
print(f'RMSPE for Gradient Boosting Regressor: {rmspe_gbr:.4f}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
import numpy as np

# Define a custom RMSPE metric
def rmspe(y_true, y_pred):
    return np.sqrt(np.mean(np.square((y_true - y_pred) / y_true)))

# Define the parameter grid for Gradient Boosting Regressor
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the Gradient Boosting Regressor
gbr = GradientBoostingRegressor(random_state=42)

# Initialize the GridSearchCV object
grid_search = GridSearchCV(estimator=gbr, param_grid=param_grid, 
                           scoring=make_scorer(rmspe, greater_is_better=False),
                           cv=5, n_jobs=-1, verbose=2)

# Fit the GridSearchCV object on the training data
grid_search.fit(X_train_scaled, y_train)

# Retrieve the best parameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print(f'Best Parameters: {best_params}')

# Predict on the test set using the best model
y_pred_gbr_best = best_model.predict(X_test_scaled)

# Calculate RMSPE for the best model
rmspe_gbr_best = rmspe(y_test, y_pred_gbr_best)
print(f'RMSPE for Best Gradient Boosting Regressor: {rmspe_gbr_best:.4f}')
```

```python
# Combine the training and test sets for full training
X_full = pd.concat([X_train, X_test])
y_full = pd.concat([y_train, y_test])

# Scale the full dataset
X_full_scaled = scaler.fit_transform(X_full)

# Train the best model on the entire dataset
best_model.fit(X_full_scaled, y_full)
```

```python
# Filter rows where 'cc_cons' is missing
df_missing = df[df['cc_cons'].isnull()]

# Prepare the features for missing data
X_missing = df_missing.drop(columns=['cc_cons', 'ID'])
X_missing_encoded = pd.get_dummies(X_missing)

# Ensure the encoded missing data matches the training data structure
X_missing_encoded = X_missing_encoded.reindex(columns=X_full.columns, fill_value=0)

# Check for any NaNs in the encoded features
print("NaNs in the encoded features before scaling:")
print(X_missing_encoded.isnull().sum().sum())
```

```python
# Fill any remaining NaNs with 0
X_missing_encoded.fillna(0, inplace=True)

# Scale the features for missing data
X_missing_scaled = scaler.transform(X_missing_encoded)

# Predict missing values
df_missing['cc_cons'] = best_model.predict(X_missing_scaled)

# Combine the data with predictions back into the original dataframe
df_combined = pd.concat([df_not_missing, df_missing])

# Display the combined dataframe
print("Combined Dataframe with Predicted Values:")
df_combined.shape
print('\n')
df_combined
```
